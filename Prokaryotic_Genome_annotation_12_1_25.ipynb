Prokka is a powerful tool for structural and functional annotation of prokaryotic genomes. It works well for both Gram-negative and Gram-positive bacteria. https://academic.oup.com/bioinformatics/article/30/14/2068/2390517

```python
!./bin/micromamba run -n prokka prokka --version
```

```python
!./bin/micromamba run -n prokka prokka \
    /content/GCF_004136375.1_ASM413637v1_genomic.fna \
    --outdir prokka_out \
    --prefix sample1
```

--locustag specifies the name of the gene you want to assign. For example, in this command, all genes will start with xoo, resulting in names like xoo1, xoo5, xoo10, etc.

--outdir defines the directory where the annotation files will be stored. Inside this directory, you will find multiple files:

The file with the .gff extension contains the complete annotation description.
Files with the .faa extension contain protein sequences.
Files with the .fna extension contain nucleotide sequences for coding sequences (CDS).  You could learn other parameters in https://github.com/tseemann/prokka

An alternative to annotate bacterial genomes is Bakta https://github.com/oschwengers/bakta

Remember: Prokka is designed only for prokaryotes. Never use Prokka for eukaryotes!
